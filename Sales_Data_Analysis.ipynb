```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/superstore.csv", encoding='latin1')
df.head()
```

```python
df.info()
df.isnull().sum()
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.describe()
```

```python
category_sales = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values('Sales', ascending=False)
category_sales
category_sales.plot(kind='bar', figsize=(8,5), title='Sales & Profit by Category')
plt.show()
```

```python
region_sales = df.groupby('Region')[['Sales']].sum().sort_values('Sales', ascending=False)
sns.barplot(x=region_sales.index, y='Sales', data=region_sales, palette='cool')
plt.title('Sales by Region')
plt.show()
```

```python
top_profit_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)
top_profit_states.plot(kind='bar', figsize=(10,5), color='teal', title='Top 10 States by Profit')
plt.show()
```

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Impact of Discount on Profit')
plt.show()
```

```python
ship_mode = df.groupby('Ship Mode')[['Sales', 'Profit']].sum()
ship_mode.plot(kind='bar', figsize=(8,5), title='Sales & Profit by Ship Mode')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Variables')
plt.show()
```
